# Gold Layer

## Gold Layer Data Loading

### Load Silver Tables
Reads products, carts, and cart items data from Silver layer.

### Purpose
Prepares data for Gold-level aggregations and analytics.

### Output
Displays datasets for further processing.

```python
products_df = spark.table("ecom_catalog.silver.products")
carts_df = spark.table("ecom_catalog.silver.carts")
cart_items_df = spark.table("ecom_catalog.silver.cart_items")

display(products_df.limit(10))
display(carts_df.limit(10))
display(cart_items_df.limit(10))
```

## Time Simulation

### Random Date Generation
Creates a random number of days for each record.

### Timestamp Creation
Generates order timestamp based on current time and random offset.

### Purpose
Simulates realistic time-based data for analysis.

### Output
Displays dataset with order timestamps.

```python
from pyspark.sql.functions import current_timestamp, rand, col
from pyspark.sql.functions import expr

carts_df = carts_df.withColumn(
    "random_days",
    (rand() * 30).cast("int")
)

carts_df = carts_df.withColumn(
    "order_timestamp",
    expr("current_timestamp() - random_days * INTERVAL 1 DAY")
).drop("random_days")

display(carts_df.limit(20))
```

## Date Generation

### Order Date Creation
Generates order_date using current date with random subtraction.

### Purpose
Creates realistic date values for time-based analysis.

### Output
Displays dataset with order date.

```python
from pyspark.sql.functions import current_timestamp, rand, date_sub

carts_df = carts_df.withColumn(
    "order_date",
    date_sub(current_timestamp().cast("date"), (rand() * 30).cast("int"))
)
display(carts_df.limit(15))
```

# KPI 1: Daily Revenue

## Daily Revenue Analysis

### Data Join
Combines cart items with order date from carts.

### Aggregation
Calculates total daily revenue.

### Purpose
Provides date-wise revenue insights.

### Output
Displays daily revenue dataset.

```python
from pyspark.sql.functions import sum,round

cart_items_with_date_df = cart_items_df.join(
    carts_df.select("cart_id", "order_date"),
    on="cart_id",
    how="left"
)

daily_revenue_df = cart_items_with_date_df \
    .groupBy("order_date") \
    .agg(
        round(sum("product_discounted_total"), 2).alias("daily_revenue")
    )

display(daily_revenue_df)
```

# KPI 2: Top Selling Products

```python
from pyspark.sql.functions import sum, desc

top_products_df = cart_items_with_date_df \
    .groupBy("product_name") \
    .agg(
        sum("quantity").alias("total_quantity"),
        round(sum("product_discounted_total"),2).alias("total_revenue")
    ) \
    .orderBy(desc("total_quantity"))

display(top_products_df.limit(20))
```

# POWER BI AGGREGATION 

## Category Sales Analysis

### Aggregation
Calculates total sales and quantity for each category.

### Purpose
Provides category-wise performance insights.

### Output
Displays aggregated category sales data.

```python
from pyspark.sql.functions import sum, round

category_sales_df = cart_items_with_date_df \
    .groupBy("category") \
    .agg(
        round(sum("product_discounted_total"), 2).alias("sales"),
        round(sum("quantity"), 2).alias("quantity")
    )

display(category_sales_df)
```

Brand Performance

## Brand Sales Analysis

### Aggregation
Calculates total sales and quantity for each brand.

### Purpose
Provides brand-wise performance insights.

### Output
Displays aggregated brand sales data.

```python
brand_sales_df = cart_items_with_date_df \
    .groupBy("brand") \
    .agg(
        round(sum("product_discounted_total"), 2).alias("sales"),
        round(sum("quantity"), 2).alias("quantity")
    )

display(brand_sales_df)
```

## Product Performance Analysis

### Aggregation
Calculates units sold, revenue, and discount for each product.

### Dimensions
Groups data by product, category, and brand.

### Purpose
Evaluates product-level performance.

### Output
Displays product performance metrics.

```python
product_perf_df = cart_items_with_date_df \
    .groupBy("product_id", "product_name", "category", "brand") \
    .agg(
        round(sum("quantity"), 2).alias("units_sold"),
        round(sum("product_discounted_total"), 2).alias("revenue"),
        round(sum("discount_value"), 2).alias("discount")
    )

display(product_perf_df)
```

## User Spending Analysis

### Aggregation
Calculates total spending and items purchased per user.

### Purpose
Provides user-level purchase behavior insights.

### Output
Displays user spending metrics.

```python
from pyspark.sql.functions import sum

user_spending_df = carts_df \
    .groupBy("user_id") \
    .agg(
        round(sum("cart_discounted_total"), 2).alias("total_spent"),
        round(sum("total_quantity"), 2).alias("items_bought")
    )

display(user_spending_df)
```

## Daily Orders Analysis

### Aggregation
Calculates total unique orders per day.

### Purpose
Provides order volume trends over time.

### Output
Displays daily order counts.

```python
from pyspark.sql.functions import countDistinct

daily_orders_df = carts_df \
    .groupBy("order_date") \
    .agg(
        round(countDistinct("cart_id"), 2).alias("total_orders")
    )

display(daily_orders_df)
```

## Average Order Value Analysis

### Aggregation
Calculates average order value per day.

### Purpose
Measures customer spending behavior over time.

### Output
Displays daily average order value.

```python
from pyspark.sql.functions import avg

aov_df = carts_df \
    .groupBy("order_date") \
    .agg(
        round(avg("cart_discounted_total"), 2).alias("avg_order_value")
    )

display(aov_df)
```

## Discount Analysis

### Aggregation
Calculates total discount given and original price for each category.

### Purpose
Analyzes discount impact on category-wise sales.

### Output
Displays discount analysis metrics.

```python
discount_analysis_df = cart_items_with_date_df \
    .groupBy("category") \
    .agg(
        round(sum("discount_value"), 2).alias("total_discount_given"),
        round(sum("product_total"), 2).alias("original_price")
    )

display(discount_analysis_df)
```

# Fact Table

## Cart Items Enrichment

### Join Operation
Combines cart items with product master data.

### Selected Fields
Includes cart details, revenue, discount, category, and brand.

### Data Formatting
Rounds numeric fields for consistency.

### Output
Displays enriched cart items dataset.

```python
from pyspark.sql.functions import col, round

cart_items_enriched_df = cart_items_df.join(
    products_df,
    on="product_id",
    how="left"
).select(
    cart_items_df["cart_id"],
    cart_items_df["product_id"],
    cart_items_df["product_name"],
    cart_items_df["price"],
    cart_items_df["quantity"],
    round(cart_items_df["revenue"], 2).alias("revenue"),
    round(cart_items_df["discount_value"], 2).alias("discount_value"),
    products_df["category"],
    products_df["brand"]
)

display(cart_items_enriched_df)
```

```python
cart_items_enriched_df = cart_items_enriched_df.join(
    carts_df.select("cart_id", "order_date"),
    on="cart_id",
    how="left"
)

display(cart_items_enriched_df)
```

## Fact Sales Table

### Data Selection
Extracts key transactional fields for analysis.

### Dimensions and Measures
Includes order date, product ID, quantity, and revenue.

### Purpose
Creates a fact table for reporting and analytics.

### Output
Displays fact sales dataset.

```python
fact_sales = cart_items_enriched_df.select(
    "order_date",
    "product_id",
    "quantity",
    "revenue"
)

display(fact_sales)
```

## Product Dimension Table

### Data Selection
Extracts product details for dimensional modeling.

### Deduplication
Removes duplicate product records.

### Purpose
Creates a product dimension table for analytics.

### Output
Displays product dimension dataset.

```python
dim_products = products_df.select(
    col("product_id"),
    col("title").alias("product_name"),
    col("category"),
    col("brand")
).dropDuplicates()

display(dim_products)
```

## Date Dimension Table

### Data Extraction
Extracts unique order dates from cart data.

### Feature Engineering
Derives year, month, and day from order_date.

### Purpose
Creates a date dimension for time-based analysis.

### Output
Displays date dimension dataset.

```python
from pyspark.sql.functions import year, month, dayofmonth

dim_date = carts_df.select("order_date").dropDuplicates() \
    .withColumn("year", year("order_date")) \
    .withColumn("month", month("order_date")) \
    .withColumn("day", dayofmonth("order_date"))

display(dim_date)
```

```python
fact_sales.write.mode("overwrite").format("delta").saveAsTable("ecom_catalog.gold.fact_sales")

dim_products.write.mode("overwrite").format("delta").saveAsTable("ecom_catalog.gold.dim_products")

dim_date.write.mode("overwrite").format("delta").saveAsTable("ecom_catalog.gold.dim_date")
```
